```python
def get_user_suggestions(data,user_id,ana):
    user_label=np.unique(data['label'][data.userId==user_id])
    user_movies_list=list(data['movieId'][data.userId==user_id])
    user_cluster_list=list(data['movieId'][(data.label.isin(user_label)) & (~data.movieId.isin(user_movies_list))])
    
    if len(user_label)==1:
        sorted_list_user=list(ana.T.sort_values(by=user_label[0],ascending=False).index)
    ## add else if the user is in 2 clusters
    return sorted_list_user

def recommend_user(data,user_id,ana):
    user_list=get_user_suggestions(data,user_id,ana)
    
    print("Based on your past views you might like the {} and {}".format(user_list[0],user_list[1]))
```
